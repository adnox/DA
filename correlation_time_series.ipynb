```python
import yfinance as yf
```

```python
tickers=["TCS.NS","ICICIBANK.NS","RELIANCE.NS","BHARTIARTL.NS","ITC.NS","MARUTI.NS","BAJFINANCE.NS"]

#downloading the ticker's data using yfinance
df=yf.download(tickers,period="3y")

#swapping the levels of column index using swaplevel
df.swaplevel(axis=1)

close=df.Close

norm=close.div(close.iloc[0]).mul(100)

ret=close.pct_change().dropna()
```

```python
ret
```

```python
ret['TCS.NS_P'] = ret['TCS.NS'].pct_change()
ret['ITC.NS_P'] = ret['ITC.NS'].pct_change()
ret.dropna(inplace=True)
ret.head()
```

```python
###moving averages TCS.N
ret['TCSmoving_avg_50'] = ret['TCS.NS'].rolling(50).mean() 
ret['TCSmoving_avg_100'] = ret['TCS.NS'].rolling(100).mean()

## moving averages ITC
ret['ITC_moving_avg_50'] =ret['ITC.NS'].rolling(50).mean()
ret['ITC_moving_avg_100'] =ret['ITC.NS'].rolling(100).mean()

print(ret)
```

```python
ret['TCS.NS'].plot()
ret.TCSmoving_avg_50.plot(label='50 period moving average')
ret.TCSmoving_avg_100.plot(label='100 period moving average')
plt.legend()
```

```python
## Rolling standard deviation TCS
ret['TCS_rolling_std'] = ret['TCS.NS_P'].rolling(100).std()

# rolling standard deviation ITC
ret['ITC_rolling_std'] = ret['ITC.NS_P'].rolling(100).std()
```

```python
ret['rolling_100_correlation'] = ret['TCS.NS_P'].rolling(100).corr(ret['ITC.NS_P'])
```

```python
ret['rolling_100_correlation']
```

```python
from matplotlib.gridspec import GridSpec

fig=plt.figure(figsize=(10,8))

Grid=GridSpec(3,2) 
```

```python
ax5=fig.add_subplot(Grid[2,:])
ax5.plot(ret.rolling_100_correlation,color='blue',
         label='rolling')
plt.axhline(y=ret['TCS.NS_P'].corr(ret['ITC.NS_P']),
            color='black',linestyle=':',
            label='constant')
ax5.set_title('Rolling correlation')
plt.legend()
```
